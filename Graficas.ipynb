# 📘 Explicación Matemática de Esperanza, Varianza y Asimetría

Se explica cómo se calcularon **esperanza (media), varianza y asimetría** para las distribuciones utilizadas en las gráficas.



## 1️⃣ Distribuciones Unimodales

### Triangular Simétrica
- Distribución: $$X \sim \text{Triangular}(a=0, b=1, c=0.5)$$ donde \(c\) es el modo relativo.
- **Esperanza (media):**  
$$
E[X] = \frac{a+b+c}{3} = \frac{0+1+0.5}{3} = 0.5
$$
- **Varianza:**  
$$
\mathrm{Var}[X] = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}
$$
- **Asimetría:**  
$$
\mathrm{Skew}[X] = 0 \quad (\text{simétrica})
$$

### Normal Estándar
- Distribución: $$X \sim N(\mu=0, \sigma^2=1)$$
- **Esperanza:** $$E[X] = \mu = 0$$
- **Varianza:** $$\mathrm{Var}[X] = \sigma^2 = 1$$
- **Asimetría:** 0 (perfectamente simétrica)



## 2️⃣ Distribuciones Asimétricas

### Triangular Asimétrica [0,1], modo 0.7
- Fórmulas similares a la triangular, pero con \(c \neq 0.5\):
$$
E[X] = \text{triang.mean()}, \quad
\mathrm{Var}[X] = \text{triang.var()}, \quad
\mathrm{Skew}[X] = \text{triang.stats(moments="s")}
$$

### Triangular [0,10], modo 7
- Modo relativo:  
$$
c_\mathrm{rel} = \frac{c-a}{b-a} = \frac{7-0}{10-0} = 0.7
$$
- Se aplican las mismas fórmulas de triangular para calcular \(E[X], \mathrm{Var}[X], \mathrm{Skew}[X]\).



## 3️⃣ Distribuciones Bimodales

### Bimodal Triangular
- PDF combinada:  
$$
p_\mathrm{total}(x) = p_1(x) + p_2(x)
$$
- Normalización:  
$$
\int p_\mathrm{total}(x) dx = 1
$$
- **Esperanza:**  
$$
E[X] = \int x\, p_\mathrm{total}(x) \, dx
$$
- **Varianza:**  
$$
\mathrm{Var}[X] = \int (x-E[X])^2\, p_\mathrm{total}(x) \, dx
$$
- **Asimetría:**  
$$
\mathrm{Skew}[X] = \frac{\int (x-E[X])^3\, p_\mathrm{total}(x) \, dx}{\mathrm{Var}[X]^{3/2}}
$$

> En Python se utilizó integración numérica con `np.trapz`.

### Bimodal Normal (mezcla)
- PDF:  
$$
p(x) = w_1 \mathcal{N}(\mu_1, \sigma_1^2) + w_2 \mathcal{N}(\mu_2, \sigma_2^2)
$$
- **Esperanza:**  
$$
E[X] = w_1 \mu_1 + w_2 \mu_2
$$
- **Varianza:**  
$$
\mathrm{Var}[X] = w_1 (\sigma_1^2 + (\mu_1 - E[X])^2) + w_2 (\sigma_2^2 + (\mu_2 - E[X])^2)
$$
- **Asimetría:**  
$$
\mathrm{Skew}[X] = 0 \quad (\text{si la mezcla es simétrica})
$$



### ✅ Resumen
- Para distribuciones estándar se usan fórmulas conocidas de probabilidad.
- Para combinaciones de PDFs (bimodales) se usa **integración numérica**:
$$
E[X] = \int x p(x) dx, \quad
\mathrm{Var}[X] = \int (x-E[X])^2 p(x) dx, \quad
\mathrm{Skew}[X] = \frac{\int (x-E[X])^3 p(x) dx}{\mathrm{Var}[X]^{3/2}}
$$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import triang, norm

# ============================================================
# 📘 GRÁFICA 1 – DISTRIBUCIONES UNIMODALES
# ============================================================

print("\n=== GRÁFICA 1: Distribuciones Unimodales ===")

# 1. Triangular simétrica en [0,1], modo en 0.5
c = 0.5
dist_tri = triang(c, loc=0, scale=1)
x = np.linspace(0, 1, 500)
pdf = dist_tri.pdf(x)

mean = dist_tri.mean()
var = dist_tri.var()
skew = dist_tri.stats(moments="s")

plt.figure(figsize=(6,4))
plt.plot(x, pdf, color="blue", label="PDF Triangular simétrica")
plt.axvline(mean, color="red", linestyle="--", label=f"E[X]={mean:.2f}")
plt.title("Distribución Triangular Simétrica [0,1]")
plt.xlabel("x"); plt.ylabel("p(x)")
plt.legend(); plt.grid(True); plt.show()

print("Triangular simétrica [0,1]:")
print(f"  Esperanza = {mean:.4f}, Varianza = {var:.4f}, Asimetría = {skew:.4f}")

# 2. Normal estándar N(0,1)
mu, sigma = 0, 1
x = np.linspace(-5, 5, 500)
pdf = norm.pdf(x, mu, sigma)

plt.figure(figsize=(6,4))
plt.plot(x, pdf, label="PDF Normal(0,1)")
plt.axvline(mu, color="red", linestyle="--", label=f"E[X]={mu}")
plt.title("Distribución Normal Estándar")
plt.xlabel("x"); plt.ylabel("p(x)")
plt.legend(); plt.grid(True); plt.show()

print("Normal estándar N(0,1):")
print(f"  Esperanza = {norm.mean(mu,sigma):.4f}, Varianza = {norm.var(mu,sigma):.4f}, Asimetría = 0")

# ============================================================
# 📘 GRÁFICA 2 – DISTRIBUCIONES ASIMÉTRICAS
# ============================================================

print("\n=== GRÁFICA 2: Distribuciones Asimétricas ===")

# 1. Triangular asimétrica en [0,1], modo en 0.7
c = 0.7
dist_tri_asym = triang(c, loc=0, scale=1)
x = np.linspace(0, 1, 500)
pdf = dist_tri_asym.pdf(x)

mean = dist_tri_asym.mean()
var = dist_tri_asym.var()
skew = dist_tri_asym.stats(moments="s")

plt.figure(figsize=(6,4))
plt.plot(x, pdf, color="green", label="PDF Triangular Asimétrica")
plt.axvline(mean, color="red", linestyle="--", label=f"E[X]={mean:.2f}")
plt.title("Distribución Triangular Asimétrica [0,1]")
plt.xlabel("x"); plt.ylabel("p(x)")
plt.legend(); plt.grid(True); plt.show()

print("Triangular asimétrica [0,1]:")
print(f"  Esperanza = {mean:.4f}, Varianza = {var:.4f}, Asimetría = {skew:.4f}")

# 2. Triangular en [0,10] con modo en 7
a, b, c = 0, 10, 7
loc, scale = a, b - a
c_rel = (c - a) / (b - a)
dist_tri_big = triang(c_rel, loc=loc, scale=scale)

x = np.linspace(a, b, 500)
pdf = dist_tri_big.pdf(x)
mean = dist_tri_big.mean()
var = dist_tri_big.var()
skew = dist_tri_big.stats(moments="s")

plt.figure(figsize=(6,4))
plt.plot(x, pdf, label="Triangular PDF")
plt.axvline(mean, color="red", linestyle="--", label=f"E[X]={mean:.2f}")
plt.axvline(c, color="green", linestyle=":", label=f"Modo={c}")
plt.title("Distribución Triangular [0,10], modo=7")
plt.xlabel("x"); plt.ylabel("p(x)")
plt.legend(); plt.grid(True); plt.show()

print("Triangular [0,10] con modo en 7:")
print(f"  Esperanza = {mean:.4f}, Varianza = {var:.4f}, Asimetría = {skew:.4f}")

# ============================================================
# 📘 GRÁFICA 3 – DISTRIBUCIONES BIMODALES
# ============================================================

print("\n=== GRÁFICA 3: Distribuciones Bimodales ===")

# 1. Bimodal triangular
def triangular_pdf(x, a, b, c):
    y = np.zeros_like(x)
    left = (x >= a) & (x < b)
    right = (x >= b) & (x <= c)
    y[left] = (x[left] - a) / ((b - a) * (c - a))
    y[right] = (c - x[right]) / ((c - b) * (c - a))
    return y

x = np.linspace(0, 10, 1000)
p1 = triangular_pdf(x, a=0, b=3, c=5)
p2 = triangular_pdf(x, a=5, b=7, c=10)
p_total = p1 + p2
p_total /= np.trapz(p_total, x)  # Normalizar

EX = np.trapz(x * p_total, x)
Var = np.trapz((x-EX)**2 * p_total, x)
Skew = np.trapz((x-EX)**3 * p_total, x) / (Var**1.5)

plt.figure(figsize=(7,4))
plt.plot(x, p_total, 'b', linewidth=2, label="PDF Bimodal Triangular")
plt.axvline(EX, color='r', linestyle='--', label=f"E[X]={EX:.2f}")
plt.title("Distribución Bimodal Triangular")
plt.xlabel("x"); plt.ylabel("p(x)")
plt.legend(); plt.grid(True); plt.show()

print("Bimodal triangular:")
print(f"  Esperanza = {EX:.4f}, Varianza = {Var:.4f}, Asimetría = {Skew:.4f}")

# 2. Bimodal normal (mezcla de dos normales)
mu1, sigma1, w1 = 3, 0.8, 0.5
mu2, sigma2, w2 = 7, 0.8, 0.5
x = np.linspace(0, 10, 500)
pdf = w1 * norm.pdf(x, mu1, sigma1) + w2 * norm.pdf(x, mu2, sigma2)
E_x = w1 * mu1 + w2 * mu2
Var = w1 * (sigma1**2 + (mu1 - E_x)**2) + w2 * (sigma2**2 + (mu2 - E_x)**2)
Skew = 0  # simétrica si w1=w2 y sigma1=sigma2

plt.figure(figsize=(7,4))
plt.plot(x, pdf, label="PDF Bimodal Normal")
plt.axvline(E_x, color="red", linestyle="--", label=f"E[X]={E_x:.2f}")
plt.axvline(mu1, color="green", linestyle=":", label=f"Modo 1={mu1}")
plt.axvline(mu2, color="blue", linestyle=":", label=f"Modo 2={mu2}")
plt.title("Distribución Bimodal (mezcla de normales)")
plt.xlabel("x"); plt.ylabel("p(x)")
plt.legend(); plt.grid(True); plt.show()

print("Bimodal normal (mezcla):")
print(f"  Esperanza = {E_x:.4f}, Varianza = {Var:.4f}, Asimetría = {Skew:.4f}")
```
